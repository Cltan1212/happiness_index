Explain: https://builtin.com/articles/feature-engineering#:~:text=Apr%2029%2C%202024-,Feature%20engineering%20is%20the%20process%20of%20selecting%2C%20manipulating%20and%20transforming,used%20in%20a%20predictive%20model.

Analyze whether health indicators (life span) are suitable for verification of happiness index: (analyzed from correlation coefficient)

There are four main functions of correlation coefficient in regression analysis:

1. Judge the relationship between the independent variable and the dependent variable to determine whether it is necessary to include the independent variable in the regression equation (if it is a simple regression, it means whether it is necessary to perform regression analysis).

2. Generally, if R is lower than ±0.5, this independent variable does not need to be included in the regression equation.

3. Use regression analysis to predict and conduct correlation analysis between the actual value and the predicted value. The correlation coefficient represents the accuracy of the regression equation, that is, the degree of fitting of the regression equation.

4. Regression analysis is one of the common methods for causal prediction, but there is a correlation between two variables, which does not necessarily mean a causal relationship. Causality is just a type of correlation.

# Machine Learning

```python
# Import
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
x = df.drop(['Country', 'SD','Ladder'], axis=1)
y = df['SD']
```

```python
# train 70%, test 30%
x_train, x_test, y_train, y_test = train_test_split(x,
                                                 y,
                                                 test_size=0.3, 
                                                 random_state = None)
```

```python
x.head()
```

```python
y.head()
```

```python
model = RandomForestRegressor(random_state=0, n_estimators=200)
model.fit(x_train, y_train)
```

```python
model_score = model.score(x_train, y_train)
print('coefficient of R^2 of the prediction:', model_score)
```

```python
# prediction
y_predicted = model.predict(x_test)
print("MSE: %.6f"% mean_squared_error(y_test, y_predicted))
print("Variance: %.6f"% r2_score(y_test, y_predicted))
```

```python
print('accuraccy of linear regression on taining set: {:.2f}'.format(model.score(x_train,y_train)))
print('accuraccy of linear regression on testing set: {:.6f}'.format(model.score(x_test,y_test)))
```

# Linear Regression Model

```python
# train, test and split again for linear regression
x = df.drop(['SD','Country','Ladder'], axis = 1)
y = df['SD']
x_train, x_test, y_train, y_test = train_test_split(x,
                                                     y,
                                                     test_size=0.3, 
                                                     random_state = 2)
```

```python
x.head()
```

```python
# Fit linear regression model
lr = LinearRegression()
lr.fit(x_train, y_train)
predict_lr = lr.predict(x_test)

print('MSE:',mean_squared_error(y_test,predict_lr))
print('RMSE:',np.sqrt(mean_squared_error(y_test,predict_lr)))
print('MAE:',mean_absolute_error(y_test,predict_lr))
print('r2 score:' , r2_score(y_test,predict_lr))

n = len(df)
p = 8
print('Adjusted_R2:',1-((1-r2_score(y_test,predict_lr))*(n-1))/(n-p-1))
```

```python
# Plot the graph
plt.figure( figsize =( 12 , 8 ))
plt.scatter(y_test,predict_lr,color='c',marker="o",s=200)
plt.xlabel('y in test')
plt.ylabel('prediction')
plt.title('Linear Regression')
```

# Random Forest Model

```python
# Fit with random forest model
rfg = RandomForestRegressor()
rfg.fit(x_train,y_train)
predict_rfg = rfg.predict(x_test)

print('MSE:',mean_squared_error(y_test,predict_rfg))
print('RMSE:',np.sqrt(mean_squared_error(y_test,predict_rfg)))
print('MAE:',mean_absolute_error(y_test,predict_rfg))
print('r2 score:' , r2_score(y_test,predict_rfg))

n = len(df)
p = 8
print('Adjusted_R2:',1-((1-r2_score(y_test,predict_rfg))*(n-1))/(n-p-1))
```

```python
# Plot the graph
plt.figure( figsize =( 12 , 8 ))
plt.scatter(y_test,predict_rfg,color='c',marker="o",s=200)
plt.xlabel('y in test')
plt.ylabel('prediction')
plt.title('Random Forest')
```

# Decision Tree Model

```python
# Fit with decision tree model
dtr = DecisionTreeRegressor()
dtr.fit(x_train,y_train)
predict_dtr = dtr.predict(x_test)

print('MSE:',mean_squared_error(y_test,predict_dtr))
print('RMSE:',np.sqrt(mean_squared_error(y_test,predict_dtr)))
print('MAE:',mean_absolute_error(y_test,predict_dtr))
print('r2 score:' , r2_score(y_test,predict_dtr))
n = len(df)
p = 8
print('Adjusted_R2:',1-((1-r2_score(y_test,predict_dtr))*(n-1))/(n-p-1))
```

```python
# Plot the graph
plt.figure( figsize =( 12 , 6 ))
plt.scatter(y_test,predict_dtr,color='c',marker="o",s=200)
plt.xlabel('y in test')
plt.ylabel('prediction')
plt.title('DecisionTreeRegressor')
```

# Conclusion

```python
y = np.array([r2_score(y_test,predict_lr),r2_score(y_test,predict_rfg),r2_score(y_test,predict_dtr)])
x = ['Linear','RandomForest','DecisionTree']

plt.figure( figsize =( 8 , 4 ))
plt.bar(x,y)
plt.title('3 regressors comparision',fontdict={'weight':'normal','size': 23,'color':'black'}) 

plt.xlabel('Regressor',fontdict={'weight': 'normal', 'size': 20,'color':'red'})
plt.ylabel('R^2_score',fontdict={'weight': 'normal', 'size': 20,'color':'blue'})
```

From the graph above, we can see that the Linear Regression has the highest $R^2$ score, which is the model that near the 70% most. Therefore, we conclude Linear Regression is the best model and will select this as our model.
